```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
from scipy.misc import derivative
```

### Уровень 0
посчитайте значение 1-ой и 2-ой производной фунции  
f(x) = x^5 + 4sin(2x) + cos(3x+3) в точке x = 1

```python
def f(x):
    return x**5 + 4*np.sin(2*x) + np.cos(3*x + 3)
```

```python
d1 = derivative(f, 1, dx=1e-6, n=1)
d2 = derivative(f, 1, dx=1e-6, n=2)
print(f'Первая производная в точке x = 1: {round(d1, 4)}')
print(f'Вторая производная в точке x = 1: {round(d2, 4)}')
```

### Уровень 1
Постройте график первой и второй производной функции  
(sin(2x+1))^5  
на отрезке [-5; 5]

```python
def f(x):
    return (np.sin(2*x + 1))**5
```

```python
new_f1 = []
x_list = np.linspace(-5,5,10000)
for i in x_list:
#     print(derivative(f, i, dx=1e-6))
    new_f1 += [derivative(f, i, dx=1e-6)]
    
fx = f(x_list)
plt.figure(figsize=(15,5))
plt.plot(x_list, fx, label='original')
plt.plot(x_list, new_f1, c='y', label='1st derivative')
plt.legend(prop={'size':15})
```

```python
new_f2 = []
x_list = np.linspace(-5,5,10000)
for i in x_list:

    new_f2 += [derivative(f, i, dx=1e-6, n=2)]
    
fx = f(x_list)
plt.figure(figsize=(15,5))
plt.plot(x_list, fx, label='original')
plt.plot(x_list, new_f2, c='r', label='2nd derivative')
plt.legend(prop={'size':15})
```
