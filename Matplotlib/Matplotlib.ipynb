```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(0, 5, 0.1)
y = x**2
```

```python
plt.plot(x,y, 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Title')
```

```python
x = range(-10,11)
y = [e**2 for e in x]


plt.figure(figsize = (13,8), dpi=100)
plt.plot(x,y, 'r')
plt.plot(y,x,)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Title')
plt.savefig('1.svg')
```

```python
x = range(-10,11)
y = [e**2 for e in x]
```
