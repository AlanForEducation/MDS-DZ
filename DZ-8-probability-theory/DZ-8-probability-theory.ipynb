Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.

```python
import scipy.stats as sts
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Нормальное распределение

**Изменение относительно параметра loc**

```python
x = np.linspace(0,20,1000)
u = 2
sigma = 0.5
l_list = []
col_name = []
for l in range(5):
    norm_rv = sts.norm(loc=u, scale=sigma)
    col_name += [f'u={u}']
    u += 4
    pdf = norm_rv.pdf(x)

    l_list += [list(pdf)]
```

```python
l_list = np.rot90(np.array(l_list))
df = pd.DataFrame(l_list, index=x[::-1], columns=col_name)
df.plot(figsize=(17,7))
```

**Изменение относительно параметра scale**

```python
x = np.linspace(0,20,1000)
u = 10
sigma = 0.5
s_list = []
col_name = []
for l in range(5):
    norm_rv = sts.norm(loc=u, scale=sigma)
    col_name += [f'sigma={round(sigma,2)}']
    sigma += 0.3
    pdf = norm_rv.pdf(x)

    s_list += [list(pdf)]
```

```python
s_list = np.rot90(np.array(s_list))
df = pd.DataFrame(s_list, index=x[::-1], columns=col_name)
df.plot(figsize=(17,7))
```

### Экспоненциальное распределение

```python
x = np.linspace(0,5,1000)
lam = 0.1
expon_list = []
col_name = []
for l in range(5):
    expon_rv = sts.expon(lam)
    col_name += [f'lam={lam}']
    lam += 0.5
    pdf = expon_rv.pdf(x)

    expon_list += [list(pdf)]
```

```python
expon_list = np.rot90(np.array(expon_list))
df = pd.DataFrame(expon_list, index=x[::-1], columns=col_name)
df.plot(figsize=(17,7))
```

### Распределение Стьюдента

```python
x = np.linspace(-10,10,1000)
n = 1
t_list = []
col_name = []
for l in range(5):
    t_rv = sts.t(n)
    col_name += [f'n={n}']
    n += 1
    pdf = t_rv.pdf(x)

    t_list += [list(pdf)]
```

```python
t_list = np.rot90(np.array(t_list))
df = pd.DataFrame(t_list, index=x[::-1], columns=col_name)
df.plot(figsize=(17,7))
```

### Задание 3
Известно, что светофор горит 54 секунды красным, 4 секунды жёлтым и 28 секунд зелёным. Посчитайте, какое количество информации несёт сообщение о цвете светофора в текущий момент. А что будет в случае 28, 28 и 28 секунд для каждого цвета?

```python
import math
```

#### В первом случае, события не равновероятны
#### Посчитаем вероятности событий:
- Светофор горит всего 54 + 4 + 28 = 86 сек
- Горит красный = 54/86
- Горит желтый = 4/86
- Горит зеленый = 28/86

```python
# Используем формулу Шеннона для подсчета информации

p1 = 54 / 86
p2 = 4 / 86
p3 = 28 / 86

I = -(p1 * math.log2(p1) + p2 * math.log2(p2) + p3 * math.log2(p3))
print(I)
```

#### В первом случае, события равновероятны
#### Посчитаем вероятности событий:
- Светофор горит всего 28 * 3 = 84 сек
- Горит красный = 28 / 84 = 1/3
- Горит желтый = 1/3
- Горит зеленый = 1/3

```python
# Используем формулу Хартли

p = 1 / 3

I = math.log2(1/p)
print(I)
```
